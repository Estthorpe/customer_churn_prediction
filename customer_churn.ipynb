```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#load the dataset
df = pd.read_csv("C:/Users/enuzo/Documents/Project Portfolio/machine_learning/customer_churn/data/telco_churn.csv")

print(df.info())
df.head()
```

```python
#handling missing values
print(df.isnull().sum())

#Drop or fill misisng values
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df['TotalCharges'] .fillna(df['TotalCharges'].median(), inplace=True)
```

```python
#Convert Categorical Variables to numerical values
df = pd.get_dummies(df, columns=[col for col in df.columns if col != 'Churn'], drop_first=True)


print(df.columns)
```

```python
# Strip spaces and unwanted characters
df.columns = df.columns.str.strip()

# Rename all columns to remove unintended modifications
df.columns = df.columns.str.replace(r'_True_True', '', regex=True)

print(df.columns)  # Check the corrected column names
```

```python
# Find the exact column that contains 'Churn'
for col in df.columns:
    if 'Churn' in col:
        df.rename(columns={col: 'Churn'}, inplace=True)
        break  # Stop after renaming the correct column

print(df.columns)  # Verify 'Churn' is now correctly named
```

```python
#Convert Churn to Binary Encoding (0&1)
# Remove leading/trailing spaces from column names

df.columns = df.columns.str.strip()

# Print unique values in 'Churn' to verify
print("Unique values in 'Churn':", df['Churn'].unique())

# Convert 'Churn' to binary (Yes -> 1, No -> 0)
df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})

# Convert to integer (ensuring no unexpected values)
df['Churn'] = df['Churn'].astype(int)

# Verify the transformation
print(df['Churn'].value_counts())
```
